```python

walmart_df.sample(300).groupby('Gender')['Purchase'].describe()
```

```python
male_sample_means=[walmart_df[walmart_df['Gender']=='M']['Purchase'].sample(300, replace=True).mean() for i in range(1000)]
```

```python
female_sample_means=[walmart_df[walmart_df['Gender']=='F']['Purchase'].sample(300, replace=True).mean() for i in range(1000)]
```

```python
sns.displot(male_sample_means, bins=100, kde=True,color='red')
sns.displot(male_sample_means, color='red',kind='kde')
plt.show()
```

```python
sns.displot(female_sample_means, bins=100, kde=True)
sns.displot(female_sample_means,kind='kde')
plt.show()
```

```python
sns.distplot(male_sample_means,label='male',bins=100, kde=True,color ='red')
sns.distplot(female_sample_means,label='female',bins=100, kde=True)
plt.legend()
plt.show()
```

```python
sns.distplot(male_sample_means,label='male',hist=False, kde=True,color ='red')
sns.distplot(female_sample_means,label='female',kde=True,hist=False)
plt.legend()
plt.show()
```

```python
male_lower_limit= np.mean(male_sample_means) - 1.96*np.std(male_sample_means)
male_upper_limit= np.mean(male_sample_means) + 1.96*np.std(male_sample_means)

#population spends for male
#(male_lower_limit,male_upper_limit)

female_lower_limit= np.mean(female_sample_means) - 1.96*np.std(female_sample_means)
female_upper_limit= np.mean(female_sample_means) + 1.96*np.std(female_sample_means)
#population spends for female
#(female_lower_limit,female_upper_limit)

#95% confidence interval to calculate the upper
#and lower limit population means using percentile method
male_confidence_interval=np.percentile(male_sample_means,[2.5,97.5])
female_confidence_interval=np.percentile(female_sample_means,[2.5,97.5])

```

```python
print("Confidence interval of male buyers: ",(male_confidence_interval))
```

```python
print("Confidence interval of female buyers: ",(female_confidence_interval))
```

```python
unmarried_sample_means=[walmart_df[walmart_df['Marital_Status']==0]['Purchase'].sample(300, replace=True).mean() for i in range(1000)]
married_sample_means=[walmart_df[walmart_df['Marital_Status']==1]['Purchase'].sample(300, replace=True).mean() for i in range(1000)]
```

```python
sns.distplot(married_sample_means,label='married',hist=False, kde=True,color ='red')
sns.distplot(unmarried_sample_means,label='unmarried',kde=True,hist=False)
plt.legend()
plt.show()
```

```python
married_lower_limit= np.mean(married_sample_means) - 1.96*np.std(married_sample_means)
married_upper_limit= np.mean(married_sample_means) + 1.96*np.std(married_sample_means)

#population spends for male
#(male_lower_limit,male_upper_limit)

unmarried_lower_limit= np.mean(unmarried_sample_means) - 1.96*np.std(unmarried_sample_means)
unmarried_upper_limit= np.mean(unmarried_sample_means) + 1.96*np.std(unmarried_sample_means)
#population spends for female
#(female_lower_limit,female_upper_limit)

#95% confidence interval to calculate the upper
#and lower limit population means using percentile method
married_confidence_interval=np.percentile(married_sample_means,[2.5,97.5])
unmarried_confidence_interval=np.percentile(unmarried_sample_means,[2.5,97.5])

```

```python
print("Confidence interval of married people: ",(married_confidence_interval))
```

```python
print("Confidence interval of unmarried people: ",(unmarried_confidence_interval))
```

## Insights and Recommendations

1. Average spending of men is almost similar to that of women but we could make an argument that the average spend is slightly  more for men given their confidence interval. Hence we could probably create an additional offer for female buyers, so that the no.of potential female buyers would increase and hence the average spend would also increase for female buyers.

2. The Average spending of married and unmarried are too close to each other. So, the differentiation between them is very low. Hence, an equal approach to target married and unmarried customers would be advised. Both couple-type products and other products will get sold based on the data given.
