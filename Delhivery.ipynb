## **Splitting source name and destination name and taking the State name alone**

```python
def getstate(col_name,del_df):      # function to retrieve state name from 
  for i in range(0,len(del_df)):    # source and destination name
    if(type(del_df.iloc[i][col_name])==str):
      dk=del_df.iloc[i][col_name].split('(')[1][:-1]
      del_df.at[i, col_name] = dk
```

# **Testing actual time aggregated value and OSRM time aggregated value**

```python
test_df=del_df[['trip_uuid','source_center','destination_center','actual_time','osrm_time']]
t2_df=test_df.groupby(['trip_uuid','source_center','destination_center']).agg({'actual_time':'last','osrm_time':'last'})
```

```python
t2_df
```

```python
# add source / destination name/center in above function. groupby with that and trip_uuid. then you'll get 2 values of last for each section. then sum it
t2_df=t2_df.reset_index()
t3_df=t2_df.groupby('trip_uuid').agg({'actual_time':'sum','osrm_time':'sum'})
t3_df
```

**Defining Null and Alternate Hypothesis for independent sample T-Test**
<br>
$H_0:$ Actual time and the OSRM calculated time are same.<br>
$H_a:$ Actual time and the OSRM calculated time are not same.<br>
Let $\mu_1$ and $\mu_2$ be the mean acutal time and mean OSRM time.
respectively.
<br>Mathematically, the above formulated hypotheses can be written as:
$H_0: \mu_1 = \mu_2$
$H_a: \mu_1 ≠ \mu_2$
<br> Significance level α = 0.05.


```python
actual_time=t3_df['actual_time'].sample(300)
osrm_time=t3_df['osrm_time'].sample(300)


from scipy.stats import ttest_ind

test_stat, p_value = ttest_ind(actual_time, osrm_time, equal_var = False, alternative = 'greater')
print('The p-value is', p_value)

if p_value < 0.05:
    print(f'As the p-value {p_value} is less than the level of significance, we reject null hypothesis')
else:
    print(f'As the p-value {p_value} is greater than the level of significance, we fail to reject null hypothesis')
```

**Hence, we could say that mean actual time and mean OSRM time are not equal**

# **Testing osrm time aggregated value and segment osrm time aggregated value**

**Defining Null and Alternate Hypothesis for independent sample T-Test**

```python
seg_osrm_time=seg_df['segment_osrm_time'].sample(300)
osrm_time=seg_df['osrm_time'].sample(300)


from scipy.stats import ttest_ind

test_stat, p_value = ttest_ind(seg_osrm_time, osrm_time, equal_var = False, alternative = 'greater')
print('The p-value is', p_value)

if p_value < 0.05:
    print(f'As the p-value {p_value} is less than the level of significance, we reject null hypothesis')
else:
    print(f'As the p-value {p_value} is greater than the level of significance, we fail to reject null hypothesis')
```
